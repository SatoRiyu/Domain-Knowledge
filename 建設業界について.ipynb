業界全体として、2007年から10年までは落ち込んでいた。  
  
2011年の震災からの復興需要、民間の設備投資、12年末からのアベノミクス（公共事業投資、景気回復）で上向きになる。  
2015年以降は、首都圏の再開発が進み、建設ラッシュが続く。東京五輪前まではこうした状況が続くと見られる。  
  
2017年の利益水準は、バブル期を超える。官公庁の発注は減ったが、民間からは増加した。  
スーパーゼネコンの新規受注高は6年連続で5兆円を超える見通しだった。（談合問題はどうなった？）  
一方、地方ゼネコンは復興工事が落ち着いてきたので、業績が伸び悩んでいる。

業界全体として、**技能労働者の高齢化**と**若手不足**が深刻化している。  


### 大手ゼネコンは、海外展開を加速している（地図が欲しい！）  
  
大林組は、北米、アジア、オセアニア、中東に進出している。  
ニュージーランドでは高速道路、  
カタールでは旧市街地の再開発、  
ラオスでは水力発電の建設を行っている。  
  
鹿島建設は、アメリカ、ヨーロッパ、東南アジア、中国、台湾、オーストラリアが拠点である。  
シンガポールの大型住宅と商業複合開発、  
ミャンマーの都市開発事業、  
ベトナムではホテルの建設を行っている。  
  
清水建設は、アメリカ、ヨーロッパ、東南アジア、中国に進出している。  
投資開発事業の一環でジャカルタでは単独開発のサービスアパートメントを建設中。  
  
竹中工務店は、アメリカ、ヨーロッパ、アジアに進出している。  
2018年9月にはインドネシアで自社設計施工による初の免震建物を完成させた。  
  
→五輪後に備えて、海外への展開を積極化している


### 五輪後も堅調の見通し、しかし人手不足
都市開発、市街地の再開発、**インフラの老朽化対策**等の需要がある。  
国内需要は、人口減少の影響で縮小する。
**新興国など海外事業を強化**している。

高齢による退職者が増える一方、若者の就職率が低く、  
2017年時点では330万人いる現場労働者は、すぐに300万人未満となる。  

機械化や作業の効率化などで生産性を向上、離職対策として高技能者の待遇を改善する試みがある。

## 自分なりにまとめると  
  
人手不足が深刻化している。  
2011年ごろからは、復興事業、アベノミクスの影響で上向きになって来ている。  
さらに、首都圏の再開発、東京五輪による開発、鉄道、高速道路等のインフラ工事で建設需要がある。  
2017年の受注高水準はバブル期を超える。  
  
五輪後も再開発、インフラ老朽化による建設需要はある。  
スーパーゼネコンは五輪後に備え、新興国等の海外需要を狙っている。  
地方ゼネコンはどういう動きなのかは調べていないので、後日、調べる。  
  
人手不足対策として、機械化、作業効率の向上、高技術者の待遇改善を行っている。
