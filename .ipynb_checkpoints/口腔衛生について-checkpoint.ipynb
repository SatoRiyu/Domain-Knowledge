### MEMS
>本来、MEMSはセンサなどの既存のデバイスの代替を主な目的として研究開発が進められていたが、近年はMEMSにしか許されない環境下での実験手段として注目されている。例えば、電子顕微鏡の中は高真空で微小な空間だが、MEMSならばその小ささと機械的性質を利用して電子顕微鏡下での実験を行うことができる。また、DNAや生体試料などのナノ・マイクロメートルの物質を操作・捕獲・分析するツールとしても活躍している。

### 唾液  
緩衝作用  
口腔内のpHを中性に調整しようとする  
嚥下効率化作用  
水分を含ませ適度な粘性を持たせる   
易消化作用(?)  
酵素により体内消化器官で消化しやすくする  
   
血液と比べ100〜1000分の一の有用な生体情報しか得られない    
  
メディカルマイクロデバイス  
人々の日常生活状況を視覚化し、未病／予防医療に貢献できる  
マイクロ基盤が電極と結合することで、マイクロチャネル内の流体が層流を形成し、微小分子を分離  
  
35歳以上の成人のうち50％以上が歯周病に罹患     
リスク予備軍まで含めると85％以上    

歯周病は、糖尿病／心臓疾患／骨粗鬆症等を増悪させる  
  
  
プロトタイプの作成
